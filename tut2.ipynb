```python
# @title Default title text

# 1. List Manipulation (L = [11, 12, 13, 14]):

L = [11, 12, 13, 14]

# (i) Add 50 and 60 to L
L.extend([50, 60])
print(L)

# (ii) Remove 11 and 13 from L
L.remove(11)
L.remove(13)
print(L)

# (iii) Sort L in ascending order
L.sort()
print(L)

# (iv) Sort L in descending order
L.sort(reverse=True)
print(L)

# (v) Search for 13 in L
if 13 in L:
    print("13 is in the list.")
else:
    print("13 is not in the list.")

# (vi) Count the number of elements present in L
print("Number of elements in L:", len(L))

# (vii) Sum all the elements in L
print("Sum of elements in L:", sum(L))

# (viii) Sum all ODD numbers in L
odd_sum = sum([x for x in L if x % 2 != 0])
print("Sum of odd numbers:", odd_sum)

# (ix) Sum all EVEN numbers in L
even_sum = sum([x for x in L if x % 2 == 0])
print("Sum of even numbers:", even_sum)

# (x) Sum all PRIME numbers in L
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

prime_sum = sum([x for x in L if is_prime(x)])
print("Sum of prime numbers:", prime_sum)

# (xi) Clear all elements in L
L.clear()
print(L)

# (xii) Delete L
del L


# 2. Sum of Items Without Inbuilt Functions:
L = [1, 2, 3, 4, 5]
total_sum = 0
for num in L:
    total_sum += num
print("Sum of items:", total_sum)


# 3. Multiply All Items Without Inbuilt Functions:
L = [1, 2, 3, 4, 5]
product = 1
for num in L:
    product *= num
print("Product of items:", product)


# 4. Generate 3D Array (3x4x6):
array = [[['*' for _ in range(6)] for _ in range(4)] for _ in range(3)]
print(array)


# 5. Dictionary Manipulation (D = {1:5.6, 2:7.8, 3:6.6, 4:8.7, 5:7.7}):
D = {1: 5.6, 2: 7.8, 3: 6.6, 4: 8.7, 5: 7.7}

# (i) Add new entry in D
D[8] = 8.8
print(D)

# (ii) Remove key=2
D.pop(2)
print(D)

# (iii) Check if key=6 is present in D
print(6 in D)

# (iv) Count the number of elements in D
print(len(D))

# (v) Add all values in D
total_value_sum = sum(D.values())
print(total_value_sum)

# (vi) Update value of key=3 to 7.1
D[3] = 7.1
print(D)

# (vii) Clear the dictionary
D.clear()
print(D)


# 6. Set Manipulation (S1 = {10, 20, 30, 40, 50, 60}, S2 = {40, 50, 60, 70, 80, 90}):
S1 = {10, 20, 30, 40, 50, 60}
S2 = {40, 50, 60, 70, 80, 90}

# (i) Add 55 and 66 to S1
S1.update([55, 66])
print(S1)

# (ii) Remove 10 and 30 from S1
S1.discard(10)
S1.discard(30)
print(S1)

# (iii) Check if 40 is present in S1
print(40 in S1)

# (iv) Union of S1 and S2
union_set = S1.union(S2)
print(union_set)

# (v) Intersection of S1 and S2
intersection_set = S1.intersection(S2)
print(intersection_set)

# (vi) S1 - S2
difference_set = S1.difference(S2)
print(difference_set)


# 7. Additional Programs:

import random
import string

# (i) Print 100 random strings whose length is between 6 and 8
for _ in range(100):
    length = random.randint(6, 8)
    rand_str = ''.join(random.choices(string.ascii_letters, k=length))
    print(rand_str)

# (ii) Prime numbers between 600 and 800
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

for i in range(600, 801):
    if is_prime(i):
        print(i)

# (iii) Numbers between 100 and 1000 divisible by 7 and 9
for i in range(100, 1001):
    if i % 7 == 0 and i % 9 == 0:
        print(i)


# 8. Display Examination Schedule:
exam_st_date = (11, 12, 2014)
print("The examination will start from: %i / %i / %i" % exam_st_date)


# 9. Iterate and Print Numbers Divisible by 5:
numbers = [10, 23, 45, 67, 100, 120]
for num in numbers:
    if num % 5 == 0:
        print(num)


# 10. Even or Odd Check Using Boolean:
num = int(input("Enter a number: "))
is_even = (num % 2 == 0)
print(f"{num} is even: {is_even}")


# 11. Count Occurrences of "Emma":
text = "Emma is a good developer. Emma works at Google. Emma is from Canada."
count = text.count("Emma")
print(f"Emma appears {count} times.")


# 12. Create New List from Two Lists (Odd from First, Even from Second):
list1 = [1, 2, 3, 4, 5, 6]
list2 = [10, 12, 15, 18, 20, 25]
new_list = [x for x in list1 if x % 2 != 0] + [x for x in list2 if x % 2 == 0]
print(new_list)
```
